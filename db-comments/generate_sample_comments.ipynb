```python
from random import randrange, choice
import datetime 
```

```python
def random_date(start,l):
   current = start
   while l >= 0:
      curr = current + datetime.timedelta(minutes=randrange(700))
      yield curr
      l-=1
```

```python
# generating a list of random dates and times
date_list = []
for i in range(40):
    for x in random_date(datetime.datetime(2019, randrange(1,12), randrange(1,28), randrange(7, 23), 00),10):
      date_list.append(x.strftime("%d/%m/%y, %H:%M:%S"))
for i in range(15): 
    for x in random_date(datetime.datetime(2020, 1, randrange(1,31), randrange(7, 23), 00),10):
      date_list.append(x.strftime("%d/%m/%y, %H:%M:%S"))
    for x in random_date(datetime.datetime(2019, 12, randrange(1,31), randrange(7, 23), 00),10):
      date_list.append(x.strftime("%d/%m/%y, %H:%M:%S"))
date_list
```

```python
# rename columns
radar_data.columns = ["technologie", "radar", "ring"]
```

```python
# function to generate comments
def add_comment(technologie, radar, ring, kommentarliste_gut, kommentarliste_schlecht, kommentarliste_evaluieren, zeitliste, autorenliste):
    zeit = zeitliste[randrange(len(zeitliste))]
    autor = autorenliste[randrange(len(autorenliste))]
    if(ring == "einsetzen"):
        meinungen = [4, 1, 2]
        meinung = meinungen[randrange(len(meinungen))]
        if(meinung == 4):
            text = kommentarliste_gut[randrange(len(kommentarliste_gut))]
        elif(meinung == 2):
            text = kommentarliste_schlecht[randrange(len(kommentarliste_schlecht))]
        elif(meinung == 1):
            text = kommentarliste_evaluieren[randrange(len(kommentarliste_evaluieren))]
    elif(ring == "überdenken"):
        meinungen = [6, 1, 3]
        meinung = meinungen[randrange(len(meinungen))]
        if(meinung == 3):
            text = kommentarliste_gut[randrange(len(kommentarliste_gut))]
        elif(meinung == 6):
            text = kommentarliste_schlecht[randrange(len(kommentarliste_schlecht))]
        elif(meinung == 1):
            text = kommentarliste_evaluieren[randrange(len(kommentarliste_evaluieren))]
    else:
        meinungen = [5, 2, 3]
        meinung = meinungen[randrange(len(meinungen))]
        if(meinung == 3):
            text = kommentarliste_gut[randrange(len(kommentarliste_gut))]
        elif(meinung == 2):
            text = kommentarliste_schlecht[randrange(len(kommentarliste_schlecht))]
        elif(meinung == 5):
            text = kommentarliste_evaluieren[randrange(len(kommentarliste_evaluieren))]
    return [autor, text, meinung, zeit, technologie, radar]
```

```python
# generate comments
kommentaranzahl = [3,4,5,6,7,8,9,14,17]
liste = []
for i, j in radar_data.iterrows():
    y=0
    while(y<choice(kommentaranzahl)):
        liste.append(add_comment(j.technologie, j.radar, j.ring, 
                                 positive_comments_list, negative_comments_list, evaluate_comments_list, date_list, names_list))
        y=y+1
```

```python
generated_comments = pd.DataFrame(liste)
```

```python
generated_comments.columns = ["autor", "text", "meinung", "zeit", "technologie", "radar"]
```
